```python
#Lets encode labels for modelling
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['label_encoded'] = le.fit_transform(df['label'])

print("Label Encoding Map")
for i, cls in enumerate(le.classes_):
    print(f"  {i} → {cls}")

print(f"\nDataset ready: {df.shape}")
print(df[['text', 'label', 'label_encoded']].head(5))
```

```python
#TEXT ClEANING PIPELINE
import re

def clean_transaction(text):
    text = str(text).upper()                        
    text = re.sub(r'\d+', '', text)                 
    text = re.sub(r'[^A-Z\s]', '', text)           
    text = re.sub(r'\s+', ' ', text).strip()        
    return text

df['text_clean'] = df['text'].apply(clean_transaction)

print("=== Before vs After Cleaning ===\n")
for i in range(5):
    print(f"Before: {df['text'][i]}")
    print(f"After:  {df['text_clean'][i]}")
    print()
```

```python
#TRAIN AND SPLIT 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df['text_clean'],
    df['label_encoded'],
    test_size=0.2,
    random_state=42,
    stratify=df['label_encoded']
)

print(f"Training samples:   {len(X_train)}")
print(f"Test samples:       {len(X_test)}")
print(f"\nClass distribution in training set:")
print(y_train.value_counts().sort_index())
```

```python
#TF & IDF VECTORIZER
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=500,
    ngram_range=(1, 2),
    analyzer='word'
)

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

print(f"TF-IDF matrix shape (train): {X_train_tfidf.shape}")
print(f"TF-IDF matrix shape (test):  {X_test_tfidf.shape}")
print(f"\nSample features learned:")
print(tfidf.get_feature_names_out()[:20])
```

```python
#Train logistic Regreession
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, f1_score
import numpy as np

lr_model = LogisticRegression(max_iter=1000, random_state=42)
lr_model.fit(X_train_tfidf, y_train)

y_pred_lr = lr_model.predict(X_test_tfidf)

print("Logistic Regression Baseline \n")
print(classification_report(y_test, y_pred_lr, 
      target_names=le.classes_))

baseline_f1 = f1_score(y_test, y_pred_lr, average='weighted')
print(f"Weighted F1 Score: {baseline_f1:.4f}")
```

```python
#Confusion Matrix
plt.figure(figsize=(9, 7))
cm = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_,
            yticklabels=le.classes_)
plt.title('Logistic Regression — Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('confusion_matrix_baseline.png', dpi=150)
plt.show()
print("Confusion matrix saved")
```

```python
def clean_wf_transaction(text):
    text = str(text).upper()
    
    # Remove the standard prefixes that add no signal
    prefixes = [
        'CHECK CRD PURCHASE', 'RECUR DEBIT CRD PMT', 
        'DEBIT CARD PURCHASE', 'POS PURCHASE',
        'ACH DEBIT', 'ONLINE TRANSFER', 'ATM WITHDRAWAL'
    ]
    for prefix in prefixes:
        text = text.replace(prefix, '')
    
    # Remove dates like 11/11, 01/23
    text = re.sub(r'\d{1,2}/\d{1,2}', '', text)
    
    # Remove card numbers (XXXXXX patterns)
    text = re.sub(r'X+', '', text)
    
    # Remove MCC codes
    text = re.sub(r'MCC=?\s*\d*', '', text)
    text = re.sub(r'\bMCC\b', '', text)
    
    # Remove phone numbers
    text = re.sub(r'\d{3}[-.\s]\d{3}[-.\s]\d{4}', '', text)
    
    # Remove US state abbreviations at end
    text = re.sub(r'\b[A-Z]{2}\b', '', text)
    
    # Remove all remaining digits
    text = re.sub(r'\d+', '', text)
    
    # Remove special characters
    text = re.sub(r'[^A-Z\s]', '', text)
    
    # Clean extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df_external['trans_clean'] = df_external['trans_desc'].apply(clean_wf_transaction)

print("=== Before vs After ===\n")
for i in range(8):
    print(f"Before: {df_external['trans_desc'][i][:80]}")
    print(f"After:  {df_external['trans_clean'][i]}")
    print()

print(f"Cleaned mean length: {df_external['trans_clean'].str.len().mean():.1f} chars")
print(f"\nEmpty strings after cleaning: {(df_external['trans_clean'] == '').sum()}")
```

```python
import torch
from torch.utils.data import Dataset

# Tokenize all transactions
def tokenize_data(texts, labels, tokenizer, max_length=64):
    encodings = tokenizer(
        list(texts),
        truncation=True,
        padding=True,
        max_length=max_length,
        return_tensors='pt'
    )
    return encodings, torch.tensor(list(labels))

# Tokenize train and test sets
print("Tokenizing training data...")
train_encodings, train_labels = tokenize_data(
    X_train, y_train, tokenizer)

print("Tokenizing test data...")
test_encodings, test_labels = tokenize_data(
    X_test, y_test, tokenizer)

print(f"\n✓ Train encodings shape: {train_encodings['input_ids'].shape}")
print(f"✓ Test encodings shape:  {test_encodings['input_ids'].shape}")
print(f"\nSample input_ids (first transaction):")
print(train_encodings['input_ids'][0])
print(f"\nSample attention_mask (first transaction):")
print(train_encodings['attention_mask'][0])
```

```python
class TransactionDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {
            'input_ids': self.encodings['input_ids'][idx],
            'attention_mask': self.encodings['attention_mask'][idx],
            'labels': self.labels[idx]
        }
        return item

# Create datasets
train_dataset = TransactionDataset(train_encodings, train_labels)
test_dataset = TransactionDataset(test_encodings, test_labels)

print(f"✓ Train dataset: {len(train_dataset)} samples")
print(f"✓ Test dataset:  {len(test_dataset)} samples")

# Verify one sample looks correct
sample = train_dataset[0]
print(f"\nSample keys: {list(sample.keys())}")
print(f"input_ids shape: {sample['input_ids'].shape}")
print(f"label: {sample['labels'].item()} → {le.classes_[sample['labels'].item()]}")
```

```python
#Load DistilBERT model
from transformers import DistilBertForSequenceClassification
from torch.utils.data import DataLoader
import torch

# Number of categories
NUM_LABELS = len(le.classes_)

# Load pretrained DistilBERT with classification head
model = DistilBertForSequenceClassification.from_pretrained(
    'distilbert-base-uncased',
    num_labels=NUM_LABELS
)

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

print(f"✓ Model loaded — {NUM_LABELS} output labels")
print(f"✓ Device: {device}")
print(f"✓ Model parameters: {sum(p.numel() for p in model.parameters()):,}")
print(f"\nOutput labels:")
for i, cls in enumerate(le.classes_):
    print(f"  {i} → {cls}")
```

```python
from torch.optim import AdamW
from transformers import get_linear_schedule_with_warmup
from sklearn.utils.class_weight import compute_class_weight
from torch.utils.data import DataLoader

# DataLoaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

print(f"✓ Train batches: {len(train_loader)}")
print(f"✓ Test batches:  {len(test_loader)}")

# Optimizer — now imported from torch not transformers
optimizer = AdamW(model.parameters(), lr=2e-5, weight_decay=0.01)

# Learning rate scheduler
EPOCHS = 3
total_steps = len(train_loader) * EPOCHS
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=total_steps // 10,
    num_training_steps=total_steps
)

# Class weights
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(device)

print(f"\n✓ Optimizer: AdamW lr=2e-5")
print(f"✓ Epochs: {EPOCHS}")
print(f"✓ Total training steps: {total_steps}")
print(f"\nClass weights (higher = more attention):")
for i, (cls, w) in enumerate(zip(le.classes_, class_weights)):
    print(f"  {cls:<25} {w:.4f}")
```

```python
#Training loop
import torch.nn as nn
from sklearn.metrics import f1_score
import time

# Loss function with class weights
loss_fn = nn.CrossEntropyLoss(weight=class_weights_tensor)

# Training function
def train_epoch(model, loader, optimizer, scheduler, loss_fn, device):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    for batch in loader:
        optimizer.zero_grad()

        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits

        loss = loss_fn(logits, labels)
        loss.backward()

        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()
        preds = torch.argmax(logits, dim=1).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(loader)
    f1 = f1_score(all_labels, all_preds, average='weighted')
    return avg_loss, f1

# Evaluation function
def evaluate(model, loader, loss_fn, device):
    model.eval()
    total_loss = 0
    all_preds, all_labels = [], []

    with torch.no_grad():
        for batch in loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits

            loss = loss_fn(logits, labels)
            total_loss += loss.item()

            preds = torch.argmax(logits, dim=1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(loader)
    f1 = f1_score(all_labels, all_preds, average='weighted')
    return avg_loss, f1, all_preds, all_labels

print("✓ Training and evaluation functions ready")
print(f"\nEstimated training time on CPU:")
print(f"  ~{len(train_loader) * EPOCHS * 2 // 60} minutes for {EPOCHS} epochs")
print(f"\nReady to train — run Cell 8 to start")
```

```python
#Run training 
print("=" * 55)
print("         FINSIGHT — BERT FINE-TUNING")
print("=" * 55)

history = {'train_loss': [], 'train_f1': [], 'val_loss': [], 'val_f1': []}
best_val_f1 = 0
best_model_state = None

for epoch in range(EPOCHS):
    start = time.time()
    print(f"\nEpoch {epoch + 1}/{EPOCHS}")
    print("-" * 40)

    # Train
    train_loss, train_f1 = train_epoch(
        model, train_loader, optimizer, scheduler, loss_fn, device)

    # Evaluate
    val_loss, val_f1, val_preds, val_labels = evaluate(
        model, test_loader, loss_fn, device)

    # Save best model
    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        best_model_state = model.state_dict().copy()
        print(f"  ★ New best model saved!")

    elapsed = time.time() - start
    history['train_loss'].append(train_loss)
    history['train_f1'].append(train_f1)
    history['val_loss'].append(val_loss)
    history['val_f1'].append(val_f1)

    print(f"  Train Loss: {train_loss:.4f} | Train F1: {train_f1:.4f}")
    print(f"  Val Loss:   {val_loss:.4f} | Val F1:   {val_f1:.4f}")
    print(f"  Time: {elapsed/60:.1f} min")

print("\n" + "=" * 55)
print(f"  Training complete — Best Val F1: {best_val_f1:.4f}")
print("=" * 55)
```

```python
def clean_transaction(text):
    text = str(text).upper()
    prefixes = [
        'CHECK CRD PURCHASE', 'RECUR DEBIT CRD PMT',
        'DEBIT CARD PURCHASE', 'POS PURCHASE',
        'ACH DEBIT', 'ONLINE TRANSFER', 'ATM WITHDRAWAL'
    ]
    for prefix in prefixes:
        text = text.replace(prefix, '')
    text = re.sub(r'\d{1,2}/\d{1,2}', '', text)
    text = re.sub(r'X+', '', text)
    text = re.sub(r'MCC=?\s*\d*', '', text)
    text = re.sub(r'\bMCC\b', '', text)
    text = re.sub(r'\d{3}[-.\s]\d{3}[-.\s]\d{4}', '', text)
    text = re.sub(r'\b[A-Z]{2}\b', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^A-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def predict_transaction(text, model, tokenizer, le, device):
    # Clean the input
    cleaned = clean_transaction(text)

    # Tokenize
    encoding = tokenizer(
        cleaned,
        truncation=True,
        padding=True,
        max_length=64,
        return_tensors='pt'
    )

    # Predict
    model.eval()
    with torch.no_grad():
        input_ids = encoding['input_ids'].to(device)
        attention_mask = encoding['attention_mask'].to(device)
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        probs = torch.softmax(logits, dim=1)
        confidence, predicted = torch.max(probs, dim=1)

    category = le.inverse_transform([predicted.item()])[0]
    confidence_pct = confidence.item() * 100

    return {
        'raw': text,
        'cleaned': cleaned,
        'category': category,
        'confidence': confidence_pct,
        'all_probs': {
            le.classes_[i]: round(probs[0][i].item() * 100, 1)
            for i in range(len(le.classes_))
        }
    }

print("✓ predict_transaction function ready")
```

```python
def clean_transaction_v2(text):
    text = str(text).upper()
    
    # Remove standard bank prefixes
    prefixes = [
        'CHECK CRD PURCHASE', 'RECUR DEBIT CRD PMT',
        'DEBIT CARD PURCHASE', 'POS PURCHASE',
        'ACH DEBIT', 'ONLINE TRANSFER', 'ATM WITHDRAWAL'
    ]
    for prefix in prefixes:
        text = text.replace(prefix, '')
    
    # Remove dates
    text = re.sub(r'\d{1,2}/\d{1,2}', '', text)
    
    # Remove card number patterns (mixed alphanumeric 4+ chars)
    text = re.sub(r'\b[A-Z0-9]*\d[A-Z0-9]*\d[A-Z0-9]*\b', '', text)
    
    # Remove X-patterns (masked card numbers)
    text = re.sub(r'X{2,}', '', text)
    
    # Remove MCC codes
    text = re.sub(r'MCC=?\s*\d*', '', text)
    text = re.sub(r'\bMCC\b', '', text)
    
    # Remove phone numbers
    text = re.sub(r'\d{3}[-.\s]\d{3}[-.\s]\d{4}', '', text)
    
    # Remove standalone numbers only (not letters mixed with numbers)
    text = re.sub(r'\b\d+\b', '', text)
    
    # Remove US state abbreviations at word boundary
    us_states = ['AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA',
                 'HI','ID','IL','IN','IA','KS','KY','LA','ME','MD',
                 'MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ',
                 'NM','NY','NC','ND','OH','OK','OR','PA','RI','SC',
                 'SD','TN','TX','UT','VT','VA','WA','WV','WI','WY']
    for state in us_states:
        text = re.sub(r'\b' + state + r'\b', '', text)
    
    # Remove special characters
    text = re.sub(r'[^A-Z\s]', '', text)
    
    # Clean whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Test the improved cleaner
test_cases = [
    "UBER TRIP 4X9K2",
    "NETFLIX MONTHLY SUB",
    "AMZN MKTP US*2K4LP",
    "SPOTIFY USA 12345",
    "WALGREENS PHARMACY",
    "FPL ENERGY BILL",
    "WALMART SUPERCENTER",
    "STARBUCKS STORE 12345"
]

print("=== Cleaning Pipeline v1 vs v2 ===\n")
for t in test_cases:
    v1 = clean_transaction(t)
    v2 = clean_transaction_v2(t)
    match = "✓" if v1 == v2 else "IMPROVED"
    print(f"Input:  {t}")
    print(f"v1:     {v1}")
    print(f"v2:     {v2}  {match}")
    print()
```

```python
# Post-processing correction layer
# Business rules for known model blind spots

CORRECTION_RULES = {
    # Streaming services — model thinks these are Utilities
    'NETFLIX': 'Entertainment',
    'SPOTIFY': 'Entertainment',
    'HULU': 'Entertainment',
    'DISNEY': 'Entertainment',
    'HBO': 'Entertainment',
    'APPLE MUSIC': 'Entertainment',
    'YOUTUBE': 'Entertainment',
    'PEACOCK': 'Entertainment',
    'PARAMOUNT': 'Entertainment',
    'TICKETMASTER': 'Entertainment',
    'AMC THEATER': 'Entertainment',

    # Ride sharing — model thinks TRIP = Food
    'UBER TRIP': 'Transport',
    'LYFT': 'Transport',

    # Pharmacies — model confuses with Shopping/Food
    'WALGREENS': 'Healthcare',
    'CVS': 'Healthcare',
    'RITE AID': 'Healthcare',
    'DUANE READE': 'Healthcare',

    # Energy/Utilities — model confuses with Transport
    'FPL': 'Utilities',
    'DUKE ENERGY': 'Utilities',
    'DOMINION': 'Utilities',
    'XCEL ENERGY': 'Utilities',
    'PSEG': 'Utilities',

    # Retail — model pulls into Food & Dining
    'WALMART': 'Shopping',
    'TARGET': 'Shopping',
    'COSTCO': 'Shopping',
    'HOME DEPOT': 'Shopping',
    'IKEA': 'Shopping',
    'BEST BUY': 'Shopping',
    'AMZN': 'Shopping',
    'AMAZON': 'Shopping',
}

def apply_correction_rules(text, predicted_category):
    text_upper = text.upper()
    for keyword, correct_category in CORRECTION_RULES.items():
        if keyword in text_upper:
            return correct_category, True  # category, was_corrected
    return predicted_category, False

def predict_transaction_v2(text, model, tokenizer, le, device):
    cleaned = clean_transaction_v2(text)
    encoding = tokenizer(
        cleaned,
        truncation=True,
        padding=True,
        max_length=64,
        return_tensors='pt'
    )
    model.eval()
    with torch.no_grad():
        input_ids = encoding['input_ids'].to(device)
        attention_mask = encoding['attention_mask'].to(device)
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        probs = torch.softmax(logits, dim=1)
        confidence, predicted = torch.max(probs, dim=1)

    raw_category = le.inverse_transform([predicted.item()])[0]
    confidence_pct = confidence.item() * 100

    # Apply correction rules
    final_category, was_corrected = apply_correction_rules(text, raw_category)

    return {
        'raw': text,
        'cleaned': cleaned,
        'category': final_category,
        'raw_category': raw_category,
        'confidence': confidence_pct,
        'corrected': was_corrected
    }

print("✓ predict_transaction_v2 with correction rules ready")
print(f"✓ {len(CORRECTION_RULES)} correction rules loaded")
```

```python
#building the Analytical engine: Spending breakdown
def generate_spending_breakdown(df):
    breakdown = df.groupby('category')['amount'].agg(
        total='sum',
        count='count',
        avg='mean'
    ).round(2)
    breakdown['percentage'] = (
        breakdown['total'] / breakdown['total'].sum() * 100
    ).round(1)
    breakdown = breakdown.sort_values('total', ascending=False)
    return breakdown

breakdown = generate_spending_breakdown(df_monthly)

print("=== FinSight — Spending Breakdown ===\n")
print(f"{'Category':<22} {'Total':>8} {'%':>6} {'Count':>6} {'Avg':>8}")
print("-" * 55)
for cat, row in breakdown.iterrows():
    print(f"{cat:<22} ${row['total']:>7.2f} {row['percentage']:>5.1f}% "
          f"{int(row['count']):>6} ${row['avg']:>7.2f}")
print("-" * 55)
print(f"{'TOTAL':<22} ${breakdown['total'].sum():>7.2f} {'100%':>6}")
```

```python
#Financial Health score
def calculate_health_score(df):
    total = df['amount'].sum()
    by_category = df.groupby('category')['amount'].sum()

    # 50/30/20 framework buckets
    needs_cats = ['Healthcare', 'Utilities', 'Transport', 'Financial Services']
    wants_cats = ['Food & Dining', 'Entertainment', 'Shopping']
    
    needs = sum(by_category.get(c, 0) for c in needs_cats)
    wants = sum(by_category.get(c, 0) for c in wants_cats)

    needs_pct = (needs / total) * 100
    wants_pct = (wants / total) * 100

    # Score components out of 100
    # Needs: ideal is ≤50%, penalise if over
    needs_score = max(0, 100 - max(0, needs_pct - 50) * 2)

    # Wants: ideal is ≤30%, penalise if over
    wants_score = max(0, 100 - max(0, wants_pct - 30) * 2)

    # Diversity score — rewarded for spending across categories
    categories_used = len(by_category)
    diversity_score = min(100, categories_used / 7 * 100)

    # Weighted final score
    final_score = (
        needs_score * 0.40 +
        wants_score * 0.40 +
        diversity_score * 0.20
    )

    return {
        'score': round(final_score),
        'needs_pct': round(needs_pct, 1),
        'wants_pct': round(wants_pct, 1),
        'needs_score': round(needs_score),
        'wants_score': round(wants_score),
        'diversity_score': round(diversity_score),
        'total_spend': total,
        'by_category': by_category.to_dict()
    }

health = calculate_health_score(df_monthly)

print("=" * 45)
print("     FINSIGHT — FINANCIAL HEALTH SCORE")
print("=" * 45)
print(f"\n  Overall Score:     {health['score']} / 100")
print(f"\n  Needs spending:    {health['needs_pct']}% of total")
print(f"  Wants spending:    {health['wants_pct']}% of total")
print(f"\n  Needs score:       {health['needs_score']} / 100")
print(f"  Wants score:       {health['wants_score']} / 100")
print(f"  Diversity score:   {health['diversity_score']} / 100")
print(f"\n  Total analyzed:    ${health['total_spend']:,.2f}")
print("=" * 45)

# Plain English feedback
print("\n=== Personalized Insights ===\n")

if health['wants_pct'] > 30:
    excess = health['wants_pct'] - 30
    print(f"⚠ Discretionary spending is {health['wants_pct']}% of budget "
          f"— {excess:.1f}% above the recommended 30%.")
    top_want = max(['Food & Dining', 'Entertainment', 'Shopping'],
                   key=lambda c: health['by_category'].get(c, 0))
    print(f"  → Your highest wants category is {top_want}. "
          f"Consider reducing by "
          f"${health['by_category'].get(top_want, 0) * 0.10:,.2f}/month.")

if health['needs_pct'] < 50:
    print(f"✓ Needs spending is well managed at {health['needs_pct']}% of budget.")

if health['score'] >= 75:
    print(f"✓ Strong overall financial health score of {health['score']}/100.")
elif health['score'] >= 50:
    print(f"→ Moderate financial health score of {health['score']}/100. "
          f"Small adjustments to discretionary spending would improve this.")
else:
    print(f"⚠ Financial health score of {health['score']}/100 suggests "
          f"a review of spending priorities.")
```

```python
#Anaomaly Detetction
def detect_anomalies(df, threshold_multiplier=2.0):
    anomalies = []
    for category in df['category'].unique():
        cat_df = df[df['category'] == category]
        if len(cat_df) < 2:
            continue
        mean_amt = cat_df['amount'].mean()
        std_amt = cat_df['amount'].std()
        threshold = mean_amt + (threshold_multiplier * std_amt)
        flagged = cat_df[cat_df['amount'] > threshold]
        for _, row in flagged.iterrows():
            anomalies.append({
                'transaction': row['transaction'],
                'amount': row['amount'],
                'category': row['category'],
                'category_avg': round(mean_amt, 2),
                'times_above_avg': round(row['amount'] / mean_amt, 1)
            })
    return pd.DataFrame(anomalies) if anomalies else pd.DataFrame()

anomalies = detect_anomalies(df_monthly)

print("=== FinSight — Anomaly Detection ===\n")
if len(anomalies) > 0:
    for _, row in anomalies.iterrows():
        print(f"⚠ {row['transaction']}")
        print(f"  Amount: ${row['amount']:,.2f} — "
              f"{row['times_above_avg']}x above category average "
              f"(${row['category_avg']:,.2f})")
        print(f"  Category: {row['category']}\n")
else:
    print("✓ No anomalies detected")
```
